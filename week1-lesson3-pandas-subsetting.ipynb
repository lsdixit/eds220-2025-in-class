## Selecting a single column

Simplest case: selecting a single column by column name.

General syntax:

`df['column_name']`

This is an example of **label-based subsetting**. Which means we select data from our data frame using the *names* of the columns, *not their position*.

### Example

````python
# Select the number of bird species observed at Mugu Lagoon in spring
mul_spring = df['MUL_spring']

mul_spring
````

````python
type(mul_spring)
````

We can do label based subsetting of a single column using attribute syntax

`df.column_name`

### Example

````python
df.MUL_spring
````

Favor `df['column_name']` instead of `df.column_name`

Why?

- if the column name matches a method/attribute
- if column name has a space or funky syntax

## Selecting multiple columns

Use a list of column names.

`df[['column1', 'column2', 'column3']]`

Notice: there are double square brackets. This is because we are passing a list of names ['column1', 'column2', 'column3'] to the selection brackets []

### Example

````python
# Select species abundance in Tijuana Estuary during winter and fall

tje_wf = df[['TJE_winter', 'TJE_fall']]
````

````python
# Check the type of tje_wf
type(tje_wf)

# Check the shape of the selection
tje_wf.shape
````

## Select using slice()

To select a slice of the columns we will use a special case of `loc` selection.

`df.loc[ : , 'column_start':'column_end']`

`column_start` and `column_end` are starting and ending points of column slice, slice includes both endpoints.

- the first value passed to `loc` is used to select rows, using a ':' as the row-selection parameter mean "select all the rows"

### Example

````python
# Select columns between CSM_winter and MUL_fall
csm_mul = df.loc[:, 'CSM_winter': 'MUL_fall']
csm_mul.head()
````

## Selecting rows

### Use a condition on the rows

`df[condition_on_rows]`

That `condition_on_rows` can be many things.

### Example

We are interested in all data after 2020

````python
# Select all rows with year > 2020
post_2020 = df[df['year'] > 2020]
post_2020
````

condition for our rows = `df['year'] > 2020` this is a `pandas.Series` with **boolean values** (True or False)

indicates which rows satisfy the condition year > 2020

````python
# Check the type of df['year'] > 2020
print(type(df['year'] > 2020))

# print the boolean series
df['year'] > 2020
````

### Check-In
Get the subset of the data frame on which the San Dieguito Wetland has at least 75 species recorded during spring.

````python
df[df['SDW_spring'] >= 75]
````

### Example

Suppose we want to look at data from 2012 to 2015 (including both years).

````python
subset = df[df['year'].between(2012, 2015)]
subset
````

- `df['year']` = column with year values, this is a `pandas.Series`
- `df['year'].between()` = between is a method for the pandas.Series. Calling it using .
- (2012, 2015): parameters. If we look at the documentation, we can see that the method includes both endpoints.
- `df['year'].between(2012, 2015)` = a pandas.Series of boolean values indicating which rows have year equal to 2012, 2013, 2014, 2015 in True/False
- when you put  `df['year'].between(2012, 2015)` in the selection brackets, we obtain the rows we need.

## Avoid using `loc` for selecting only rows.

It is equivalent to write

```
# Select rows with year < 2015
df[df['year'] < 2015]
```

and

```
df.loc[df['year'] < 2015, :]
```

In the second one, we are using `df.loc[row-selection, column-selection]`. the row selection is the condition `df['year'] < 2015` and the column selection is `:` which is all columns.

## using multiple conditions
We can combine conditions to select rows by surrounding each in parentheses `()` and using the or operator `|` and the and operator `&`

````python
# Select rows in CSM with more than 50 species in winter and fall
df[(df['CSM_winter'] > 50) | (df['CSM_fall'] > 50)]
````

Select rows in which both CSM and SDW have more than 60 bird species during spring

````python
df[(df['CSM_spring'] > 60) & (df['SDW_spring'] > 60)]
````

````python
cond1 = df['CSM_winter'] > 50
cond2 = df['CSM_fall'] > 50
df[cond1 | cond2]
````

### Selecting rows by position

Sometimes we may want to select rows based on their *actual position* in the data frame. 
This is the **position-based subsetting**. We use `iloc` selection.

Syntax:
```
df.iloc[row-indices]
```
`iloc` = integer 

````python
df.iloc[4]
````

## Select rows and columns simultaneously
Can be done using `iloc` (labels) or `iloc` (positions)

### by labels or conditions
Syntax: `df.loc[row selection, column selection]`

### Example
Select winter surveys for Mugu lagoon and tijuana estuary after 2020

````python
df.loc[df['year']>2020, ['MUL_winter', 'TJE_winter']]
````

### Select by position
Use `iloc`

Syntax: 
```
df.iloc[row-indices, column-indices]
```

### Example
Suppose we want to select rows 3-7 (including 7) and columns 4 and 5


````python
df.iloc[3:8, [3,4]]
````
